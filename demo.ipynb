# 生成 EEG 模拟信号

```
import numpy as np
import matplotlib.pyplot as plt

def generate_eeg(freq, duration=1.0, fs=256, noise_level=0.5):
    t = np.linspace(0, duration, int(fs * duration), endpoint=False)
    signal = np.sin(2 * np.pi * freq * t)
    noise = noise_level * np.random.randn(len(t))
    return t, signal + noise

# 模拟 8Hz 和 12Hz 信号
t, eeg_8hz = generate_eeg(8)
_, eeg_12hz = generate_eeg(12)

# 可视化前1秒信号
plt.figure(figsize=(10, 4))
plt.plot(t, eeg_8hz, label='8Hz EEG')
plt.plot(t, eeg_12hz, label='12Hz EEG', alpha=0.7)
plt.title("Simulated EEG Signal (8Hz vs. 12Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

# 提取 FFT 特征并可视化频谱

```
def compute_fft(signal, fs=256):
    fft_vals = np.abs(np.fft.rfft(signal))
    freqs = np.fft.rfftfreq(len(signal), 1/fs)
    return freqs, fft_vals

# 计算两条信号的频谱
freqs, fft_8hz = compute_fft(eeg_8hz)
_, fft_12hz = compute_fft(eeg_12hz)

# 可视化 0~30Hz 频段
plt.figure(figsize=(10, 4))
plt.plot(freqs, fft_8hz, label='8Hz EEG')
plt.plot(freqs, fft_12hz, label='12Hz EEG', alpha=0.7)
plt.xlim(0, 30)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("Frequency Spectrum of Simulated EEG")
plt.legend()
plt.show()
```

# 批量生成样本 + 特征提取

```
# 批量生成模拟 EEG 样本
def create_dataset(n_samples=400, fs=256):
    X, y = [], []
    for _ in range(n_samples):
        if np.random.rand() > 0.5:
            _, sig = generate_eeg(8, fs=fs)
            y.append(0)
        else:
            _, sig = generate_eeg(12, fs=fs)
            y.append(1)
        X.append(sig)
    return np.array(X), np.array(y)

# 提取 6~14Hz 区间的频域功率特征
def extract_band_power(X, fs=256, band=(6, 14)):
    X_feat = []
    for sig in X:
        freqs, fft_vals = compute_fft(sig, fs)
        band_mask = (freqs >= band[0]) & (freqs <= band[1])
        X_feat.append(fft_vals[band_mask])
    return np.array(X_feat)

# 构建样本与特征
X_raw, y = create_dataset()
X_feat = extract_band_power(X_raw)
print(f"X_feat shape: {X_feat.shape}, y shape: {y.shape}")
```

# 构建并训练 MLP 分类模型（使用 PyTorch）

```
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 数据准备
X_train, X_test, y_train, y_test = train_test_split(X_feat, y, test_size=0.2, random_state=42)

# 转为 tensor
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.long)
y_test = torch.tensor(y_test, dtype=torch.long)

# 定义模型
class EEGClassifier(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 2)  # 输出2类
        )

    def forward(self, x):
        return self.net(x)

model = EEGClassifier(input_dim=X_train.shape[1])

# 损失函数和优化器
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

loss_list = []

# 训练模型
for epoch in range(50):
    model.train()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    loss_list.append(loss.item())

    if epoch % 10 == 0:
        model.eval()
        preds = model(X_test).argmax(dim=1)
        acc = accuracy_score(y_test.numpy(), preds.numpy())
        print(f"Epoch {epoch:2d} | Loss: {loss.item():.4f} | Test Accuracy: {acc:.2f}")

# 绘制 Loss 曲线
import matplotlib.pyplot as plt

plt.plot(loss_list)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

频谱热图（400个信号的频域特征）

```
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(X_feat, cmap='viridis', xticklabels=np.arange(6,15))
plt.title("EEG Band Power (6–14 Hz) Across All Samples")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Sample Index")
plt.show()
```

降维后绘制决策边界

```
from sklearn.decomposition import PCA

# 降维
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_feat)

# 模型输出预测
with torch.no_grad():
    pred_label = model(torch.tensor(X_feat, dtype=torch.float32)).argmax(dim=1).numpy()

# 绘图
plt.figure(figsize=(8,6))
plt.scatter(X_2d[:,0], X_2d[:,1], c=pred_label, cmap='coolwarm', alpha=0.7)
plt.title("PCA Projection of EEG Band Features (Colored by Predicted Class)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.colorbar(label='Predicted Class')
plt.show()
```
